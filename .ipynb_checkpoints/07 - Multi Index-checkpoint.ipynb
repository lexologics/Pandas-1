# Extract Index Level Values with the <i>get_level_values</i> Method

# Change Index Level Name with the <i>set_names</i> Method

# The <i>transpose</i> Method

```python
bigmac = pd.read_csv('data/bigmac.csv', parse_dates = ['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac = bigmac.transpose()
bigmac.head(3)
```

```python
bigmac.loc[('Price in US Dollars',)]
```

```python
bigmac.loc[('Price in US Dollars',), ('2010-01-01',)]
```

```python
bigmac.loc[('Price in US Dollars',), ('2010-01-01','Sri Lanka')]
```

```python
bigmac.loc[('Price in US Dollars',), ('2010-01-01','Sri Lanka'):('2010-01-01', 'Ukraine')]  # slicing
```

# The <i>swaplevel</i> Method

```python
bigmac = pd.read_csv('data/bigmac.csv', parse_dates = ['Date'], index_col = ['Date', 'Country'])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.swaplevel('Date', 'Country')
```

```python
bigmac.swaplevel('Country', 'Date')
```

```python
bigmac = bigmac.swaplevel(0, 1)
```

# The <i>.unstack</i> Method, Part 3

```python
world = pd.read_csv('data/worldstats.csv', index_col = ['country', 'year'])
s = world.stack()
s.head(3)
```

```python
s.unstack(level = [1, 0])
```

```python
s.unstack(level = [0, 1])
```

```python
s.unstack(level = ['year', 'country'])
```

```python
s.unstack('year', fill_value = 0)  # Remove all NaN and replace with '0' (zero)
```

```python
s = s.unstack('year', fill_value = 0)
s.head(3)
```

# The <i>pivot</i> Method

```python
sales = pd.read_csv('data/salesmen.csv', parse_dates = ['Date'])
sales['Salesman'] = sales['Salesman'].astype('category')
sales.head(3)
```

```python
sales.pivot(index = 'Date', columns = 'Salesman', values = 'Revenue')
```

# The <i>pivot_table</i> Method (Excel Method)

```python
foods = pd.read_csv('data/foods.csv')
foods.head(3)
```

```python
foods.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'mean')
```

```python
foods.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'sum')
```

```python
foods.pivot_table(values = 'Spend', index = 'Item', aggfunc = 'sum')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], aggfunc = 'sum')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], columns = ['Frequency', 'City'], aggfunc = 'sum')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], columns = 'City', aggfunc = 'count')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], columns = 'City', aggfunc = 'max')
```

```python
foods.pivot_table(values = 'Spend', index = ['Gender', 'Item'], columns = 'City', aggfunc = 'min')
```

```python
pd.pivot_table(data = foods, values = 'Spend', index = ['Gender', 'Item'], columns = 'City', aggfunc = 'min')
```
