# Filter with More than One Condition (AND)

```python
df = pd.read_csv('data/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df['Team'] = df['Team'].astype('category')
df.head(3)
```

```python
mask1 = df['Gender'] == 'Male'
mask2 = df['Team'] == 'Marketing'
```

```python
df[mask1 & mask2].head()
```

# Filter with More than One Condition (OR)

```python
df = pd.read_csv('data/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df['Team'] = df['Team'].astype('category')
df.head(3)
```

```python
mask1 = df['Senior Management']
mask2 = df['Start Date'] < '1990-01-01'
df[mask1 | mask2]
```

```python
mask1 = df['First Name'] == 'Robert'
mask2 = df['Team'] == 'Client Services'
mask3 = df['Start Date'] > '2016-06-01'
df[(mask1 & mask2) | mask3]
```

# The <i>.between()</i> Method

```python
df = pd.read_csv('data/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df['Team'] = df['Team'].astype('category')
df.head(3)
```

```python
df[df['Salary'].between(60000, 70000)]
```

```python
df[df['Bonus %'].between(2.0, 5.0)]
```

```python
df[df['Start Date'].between('1991-01-01', '1992-01-01')]
```

```python
df[df['Last Login Time'].between('08:30AM', '12:00PM')]
```

# The <i>.duplicated()</i> Method

```python
df = pd.read_csv('data/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df['Team'] = df['Team'].astype('category')
df.sort_values('First Name', inplace=True)
df.head(3)
```

```python
df[df['First Name'].duplicated()]
```

```python
df[df['First Name'].duplicated(keep='last')]
```

```python
df[df['First Name'].duplicated(keep=False)]
```

```python
mask = ~df['First Name'].duplicated(keep= False)  # ~ let see all the unique rows
df[mask]
```

# The <i>.drop_duplicates()</i> Method

```python
df = pd.read_csv('data/employees.csv', parse_dates=['Start Date', 'Last Login Time'])
df['Senior Management'] = df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df['Team'] = df['Team'].astype('category')
df.sort_values('First Name', inplace=True)
df.head(3)
```

```python
len(df)
```

```python
len(df.drop_duplicates())
```

```python
df.drop_duplicates(subset = ['First Name'], keep = False)
```
