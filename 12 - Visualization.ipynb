# Bar Charts

```python
bb = data.DataReader(name='AAPL', data_source='yahoo', start='2016-01-01', end='2020-01-01')
bb.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return 'Poor'
    elif stock_price > 200 and stock_price <= 220:
        return 'Satisfactory'
    else:
        return 'Stellar!!'
```

```python
bb['Close']
```

```python
bb['Close'].apply(rank_performance)
```

```python
bb['Close'].apply(rank_performance).value_counts()
```

```python
bb['Close'].apply(rank_performance).value_counts().plot()
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind='line') # the same as abond line
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind='bar')
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind='barh')
```

# Pie Charts

```python
bb = data.DataReader(name='AAPL', data_source='yahoo', start='2016-01-01', end='2020-01-01')
bb.head(3)
```

```python
bb['Close'].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 163.10282307922483:
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
plt.style.use('ggplot')
bb['Close'].apply(rank_performance)
```

```python
plt.style.use('ggplot')
bb['Close'].apply(rank_performance).value_counts()
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind='pie')
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind='pie', legend=True)
```

# Histograms

```python
bb = data.DataReader(name='AAPL', data_source='yahoo', start='2016-01-01', end='2020-01-01')
bb.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
bb['High'].apply(custom_round).value_counts().sort_index()
```

```python
plt.style.use('ggplot')
bb['High'].apply(custom_round).value_counts().plot()
```

```python
bb['High'].apply(custom_round).nunique()
```

```python
bb['High'].apply(custom_round).plot(kind='hist')
```

```python
bb['High'].apply(custom_round).plot(kind='hist', bins=3)
```
